```python
import pandas as pd
```

```python
meta_train = pd.read_csv('meta_data_testing_parsed.csv')
meta_train
```

```python
meta_train_encoded = pd.get_dummies(meta_train, columns = ['Treatment'])
meta_train_encoded.columns
```

```python
meta_train_encoded
```

```python
meta_train_encoded.to_csv('meta_data_testing_parsed_encoded_imputed.csv', index = False)
```

## KNN Imputation

```python
# drop Env column
env = meta_train_encoded[['Env']]
meta_train_encoded_noenv = meta_train_encoded.drop(columns = 'Env')
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5, weights="uniform")
imputer.fit_transform(meta_train_encoded_noenv)
imputed_meta_train = pd.DataFrame(imputer.transform(meta_train_encoded_noenv), index=meta_train_encoded_noenv.index, columns=meta_train_encoded_noenv.columns)
imputed_meta_train
```

```python
pd.concat([env, imputed_meta_train], axis=1).to_csv('meta_data_training_parsed_encoded_imputed.csv', index = False)
```
